```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
num_samples = 500
X = np.random.uniform(-1,1, size=(num_samples,2))
```

```python
y = np.logical_xor(X[:,0]>0, X[:,1]> 0)
y = np.where(y,1,-1)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('matplot exercise')
plt.show()
```
