```python
import numpy as np
```

```python
from scipy import signal
```

```python
from scipy import misc
```

```python
import matplotlib.pyplot as plt
```

```python
image = misc.ascent()
```

```python
image
```

```python
plt.imshow(image)
```

```python
Hx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
```

```python
Hy = np.array([[-1,2,-1],[0,0,0],[1,2,1]])
```

```python
Gx = signal.convolve2d(image, Hx, mode='same')
```

```python
Gy = signal.convolve2d(image, Hy, mode='same')
```

```python
Hx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
```

```python
Hy = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
```

```python
Gx = signal.convolve2d(image, Hx, mode='same')
```

```python
Gy = signal.convolve2d(image, Hy, mode='same')
```

```python
G = np.sqrt(Gx**2 + Gy ** 2)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(8,4))
```

```python
ax1.imshow(image, cmap=plt.cm.gray)
```

```python
ax1.axis('off')
```

```python
ax1.set_title('Original Image')
```

```python
ax2.imshow(G, cmap=plt.cm.gray)
```

```python
ax2.axis('off')
```

```python
ax2.set_title('Edge-Detected Image')
```

```python
plt.show()
```
